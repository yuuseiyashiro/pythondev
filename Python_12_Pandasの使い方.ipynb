### 特定の条件による抽出

#### `.query()`メソッドを使う方法

```python
# 世帯数が30未満の地域
df.query("households < 30")
```

#### 真偽値で指定する方法

```python
# 人口が1000人以下か
df["population"] <= 1000
```

```python
# 人口が1000人以下の地域
df[df["population"] <= 1000]
```

#### 複数の条件を指定する

##### `query()` の場合

```python
df.query("population > 100 and households < 30")
```

```python
df.query("population < 100 or households < 30")
```

##### `query()` を使わない方法による複数条件の指定

```python
df[(df["population"] > 100) & (df["households"]  < 30)]
```

```python
df[(df["population"] < 100) | (df["households"]  < 30)]
```

## 列を更新・追加する

### データ型の変更

```python
df
```

```python
df[['households', 'population']] = df[['households', 'population']].astype(int)
df[['households', 'population']]
```

```python
df
```

```python
df["median_income"] = df.median_income.apply(round)
df[["median_income"]]
```

### 列の追加

```python
df["households_per_capita"] = df.households / df.population
df[["households_per_capita"]]
```
